# Variables and Data Types 

A variable in Python is a name that is used to store a value in memory. It acts as a container that holds data which can be used and modified throughout a program.


When you assign a value to a variable, Python automatically creates the variable and stores the value in it — you don’t need to declare its type explicitly.

variable_name = value


```python
x = 10
name = "Mohan"
```

Here,

x is a variable storing an integer value 10.

name is a variable storing a string value "Mohan".

Rules for Naming Variables
1.	Must start with a letter or underscore (_).
2.	Cannot start with a number.
3.	Can contain letters, digits, and underscores.
4.	Are case-sensitive (Name and name are different).
5.	Cannot use Python keywords (for, if, class, etc.)
